```python
import requests
from bs4 import BeautifulSoup
import csv
import datetime
import re
from time import sleep

#changing request header from python to browser
session = requests.Session()
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.2 (KHTML, like Gecko) Chrome/22.0.1216.0 Safari/537.2'}

#{'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux i686; rv:15.0) Gecko/20100101 Firefox/15.0.1'}

page_number = 0

def google_scraper(escaped_search_term, page_number):
       
    target_url = "https://www.google.com/search?q={}&num=100&start={}&tbs=qdr:w".format(escaped_search_term, page_number)
    print(target_url)
    
    req = session.get(target_url, headers=headers, timeout=10)
    bs = BeautifulSoup(req.text, 'html.parser')
    counter = 1
    #scrape for one page
    for post in bs.find_all('div', attrs={'class' : 'g'}):  
        title = post.find('h3').text
        links = str(post.find('a', href=True))
        stripped_link = re.search(r'"([^"]*)"', links)

        try:
            description = post.find('span', class_='f').next_sibling.strip()
        except:
            description = "No description found"
            continue
        
        post_line = [counter, title, (stripped_link.group(0)).replace('"', ''), description]
        
        with open('output.csv', 'a', encoding='UTF-8') as f:
            writer = csv.writer(f)
            writer.writerow(post_line)
        
        print(post_line)
        counter +=1
    ############
    return counter
    sleep(15)
    print("completed page number " + str(page_number))
```

```python
search_item = input("What is your search term? ")
escaped_search_term = search_item.replace(' ', '+')

for n in range(10):
    google_scraper(escaped_search_term, n)    
```
